```python
d = {'col1': [1,2,4,3,2,1,6,7,3,9], 'col2': [5,8,9,10,12,8,6,2,8,7]}
df2 = pd.DataFrame(d)
df2
```

```python
fig = plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(df2['col1'],'k.',label = 'col 1')
ax.plot(df2['col2'],'ko--', label='col 2')
ax.legend(loc='best')
```

```python
fig = plt.figure()
ax =fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum())
```

```python
ticks = ax.set_xticks([0,250,500,750,1000])
labels = ax.set_xticklabels(['one','two','three','four','five'], rotation=30, fontsize='small')
ax.set_title('My First Matploylib plot with ticks')
ax.set_xlabel('Stages')
ax.set_ylabel('Values')

fig
```

```python
props = {'title': 'My first matplotlip plot with ticks',
        'xlabel': 'Stages',
        'ylabel': 'Values'}
ax.set(**props)
fig
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(randn(1000).cumsum(), 'k', label='One')
ax.plot(randn(1000).cumsum(), 'k--', label='Two')
ax.plot(randn(1000).cumsum(), 'k.', label='Three')

ax.legend(loc='best')
```
